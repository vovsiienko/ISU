```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
ITSkillsDF=pd.read_csv("/content/drive/MyDrive/IT496/ITSkills.csv")
print(ITSkillsDF)
```

```python
ITSkillsDF.index = ITSkillsDF['Skills']
ITSkillsDF = ITSkillsDF.sort_values(by='No of people', ascending=False)

print(ITSkillsDF)
```

```python
ITSkillsDF['cumulativePercentage'] = ITSkillsDF['No of people'].cumsum()/ITSkillsDF['No of people'].sum()*100

ITSkillsDF['cumulativePercentage'] = ITSkillsDF['cumulativePercentage'].apply(lambda x: round(x,2))

print(ITSkillsDF)
```

```python
fig, ax = plt.subplots()
ax.bar(ITSkillsDF.index, ITSkillsDF['No of people'], color="C0")

ax.set_xticklabels(ITSkillsDF['Skills'], rotation=60)

ax2 = ax.twinx()

ax2.plot(ITSkillsDF.index, ITSkillsDF['cumulativePercentage'], color="C1", marker="D", ms=7, label=ITSkillsDF['cumulativePercentage'])
ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis="y", colors="C0")
ax2.tick_params(axis="y", colors="C1")

for index, v in ITSkillsDF.iterrows():
  label = round(v['cumulativePercentage'],1)
  plt.annotate(label, xy=(v['Skills'],v['cumulativePercentage']))
  
plt.show()
```
