```python
#import a few libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib as plt

#Create a dataframe (data table)

securityRisksDF=pd.read_csv("/content/drive/MyDrive/IT496/SecurityRisks.csv")
#securityRisksDF = securityRisksDF.sort_values(by='Score', ascending=False)
print(securityRisksDF)
```

```python
securityRisksDF = pd.DataFrame(columns=['Risk', 'Probability', 'Impact', 'Score'])
securityRisksDF['Risk']=['Ransomware attack', 'Data leak', 'DDoS attack', 'Phishing', 'Insider threat']
securityRisksDF['Probability']=[0.4,0.5,0.2,0.3,0.1]
securityRisksDF['Impact']=[10,9,8,7,6]
securityRisksDF['Score']=securityRisksDF['Impact']*securityRisksDF['Probability']

scoresPivot=securityRisksDF.pivot('Impact', 'Probability', 'Score')
#print(scoresPivot)
labelsPivot=securityRisksDF.pivot('Impact', 'Probability','Risk')
#print(labelsPivot)
```

```python
#simple heatmap
sns.set(rc={'figure.figsize':(14,7)})
p1=sns.heatmap(scoresPivot, cmap="flare", annot=labelsPivot, fmt='')
```
